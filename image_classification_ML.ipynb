```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import resize
# resize to make images even
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
# adding images to the file
data_array=[]
# whenever we are adding images if its cat 1 is added if its dog 0 is added
target_array=[]
categories=['Cat','Dog']
datadir='/content/drive/MyDrive/datasets/Data'
```

```python
img=imread('/content/drive/MyDrive/datasets/Data/Cat/cat.11.jpg')
img
```

```python
for i in categories:
  sub_path=os.path.join(datadir,i) # joins cat and dog
  # print(sub_path)
  images=os.listdir(sub_path) # makes it as lists
  for img in images:
    img_path=os.path.join(sub_path,img) # in for loop, print each images in sub_path
    # print(img_path)
    img_array=imread(img_path) # read the images
    img_resize=resize(img_array,(150,150,3)) # resize all the images
    data_array.append(img_resize.flatten()) #flattened means to change to 1-dimension
    target_array.append(categories.index(i)) # when each time the loop executes, target array gets updated as 0 or 1
  print('loaded category : ',i)
```

```python
import numpy as np
```

```python
data=np.array(data_array)
data.shape # 150*150*3=67500 columns , 126 images
```

```python
df=pd.DataFrame(data)
df
#all the values has been scaled while resizeing it.
```

```python
target=np.array(target_array)
target.shape
```

```python
df['target']=target
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
sv=SVC()
sv.fit(X_train,y_train)
y_pred=sv.predict(X_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
